```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
import os
os.getcwd()

```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
# Check shape and column names
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# See info about data types and nulls
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# To check if there are missing values:
df = df.ffill()  # Forward fill 
df.isnull().sum()  # Recheck

```

```python
# Normalize math, reading, and writing scores between 0 and 1
scaler = MinMaxScaler()
df[['math score', 'reading score', 'writing score']] = scaler.fit_transform(
    df[['math score', 'reading score', 'writing score']]
)

df.head()
```

```python
# Show unique values in each categorical column
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].unique()}")

# Label encode for test preparation course
le = LabelEncoder()
df['test preparation course'] = le.fit_transform(df['test preparation course'])  # completed=1, none=0

# Encode the rest of the categorical features
df = pd.get_dummies(df, columns=[
    'gender',
    'race/ethnicity',
    'parental level of education',
    'lunch'
], drop_first=True)

df.head()
```
